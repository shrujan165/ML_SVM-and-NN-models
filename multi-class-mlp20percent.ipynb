```python
def create_better_features(df):
    df = df.copy()
    
    
    df['total_engagement'] = (
        df['hobby_engagement_level'] + 
        df['physical_activity_index'] + 
        df['creative_expression_index'] + 
        df['altruism_score']
    )
    df['support_vs_consistency'] = df['support_environment_score'] / (df['consistency_score'] + 1)
    return df


def drop_original_features(df):
    features_to_drop = [
        'hobby_engagement_level', 
        'physical_activity_index', 
        'creative_expression_index', 
        'altruism_score'
    ]
    return df.drop(columns=features_to_drop, errors='ignore')

if not df_train.empty:
    # Extract IDs and Target
    test_ids = df_test['participant_id']
    y_train_full = df_train['personality_cluster']

    # Encode Target
    label_encoder = LabelEncoder()
    y_encoded_full = label_encoder.fit_transform(y_train_full)

    # Prepare Features
    X_train_fe = create_better_features(df_train.drop(columns=['participant_id', 'personality_cluster']))
    X_test_fe = create_better_features(df_test.drop(columns=['participant_id']))

    X_train_final = drop_original_features(X_train_fe)
    X_test_final = drop_original_features(X_test_fe)

    # MODIFICATION: SELECT 20% OF DATA 
    # Using stratified split to ensure all personality clusters are represented
    X_train_20, _, y_train_20, _ = train_test_split(
        X_train_final, 
        y_encoded_full, 
        train_size=0.80, 
        random_state=42, 
        stratify=y_encoded_full
    )
    
    print(f"Data shape after reducing to 20%: {X_train_20.shape}")

   
    
    # Define feature groups based on the new X_train_final columns
    # Note: Ensure these columns exist in the processed dataframe
    polynomial_features = ['focus_intensity', 'consistency_score']
    standard_scaled_features = ['total_engagement', 'support_vs_consistency']
    
    # Categorical/Ordinal columns (ensure these remain in X_train_final)
    ordinal_features = [
        'age_group', 'identity_code', 'cultural_background', 
        'upbringing_influence', 'external_guidance_usage', 'support_environment_score'
    ]

    poly_scaler_pipeline = Pipeline([
        ('poly', PolynomialFeatures(degree=2, include_bias=False)),
        ('scaler', StandardScaler())
    ])

    # ColumnTransformer handles all remaining feature types
    preprocessor_robust = ColumnTransformer(
        transformers=[
            ('poly_num', poly_scaler_pipeline, polynomial_features),
            ('std_num', StandardScaler(), standard_scaled_features),
            ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ordinal_features)
        ],
        remainder='drop'
    )

    
    
    nn_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor_robust),
        ('mlp', MLPClassifier(random_state=42, max_iter=1000))
    ])

    param_grid = {
        'mlp__hidden_layer_sizes': [(100,), (128, 64), (200, 100), (100, 50, 25)],
        'mlp__activation': ['relu', 'tanh','logistic'],
        # Reduced parameter space slightly for faster demonstration on 20% data
        'mlp__alpha': [0.00001, 0.0001,0.01,0.1] 
    }

    
    
    grid_search = GridSearchCV(
        estimator=nn_pipeline,
        param_grid=param_grid,
        scoring='f1_macro', 
        cv=5,
        verbose=2,
        n_jobs=-1
    )

    print("Starting training on 20% of the data...")
    # Fit on the 20% subset
    grid_search.fit(X_train_20, y_train_20)

    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    best_model = grid_search.best_estimator_

    print(f"Optimal Params: {best_params}")
    print(f"Best Cross-Validation Macro F1 Score (on 20% data): {best_score:.4f}")

    # --- 7. Prediction and Submission ---
    
    # Predict on the full test set using the model trained on 20% of data
    test_predictions_encoded = best_model.predict(X_test_final)
    test_predictions_cluster = label_encoder.inverse_transform(test_predictions_encoded)

    submission_df_optimized = pd.DataFrame({
        'participant_id': test_ids,
        'personality_cluster': test_predictions_cluster
    })

    submission_file_path = 'submission.csv'
    submission_df_optimized.to_csv(submission_file_path, index=False)

    print("Submission file created successfully.")
```
